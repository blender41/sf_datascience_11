Создайте с помощью функции-преобразования новый признак **"Образование"**, который должен иметь 4 категории: "высшее", "неоконченное высшее", "среднее специальное" и "среднее".

```python
#ваш код здесь
def extract_education(edu_string):
    level = " ".join(edu_string.split()[:2]).lower()  # Берем первые два слова и приводим к нижнему регистру
    
    if "высшее" in level:
        return "высшее"
    elif "неоконченное высшее" in level:
        return "неоконченное высшее"
    elif "среднее специальное" in level:
        return "среднее специальное"
    elif "среднее" in level:
        return "среднее"
    else:
        return None

# Создаем столбец "Образование"
df["Образование"] = df["Образование и ВУЗ"].apply(extract_education)

# проверяем какие уникальные категории получились
print(df["Образование"].unique())

# удаляем старый столбец
df.drop(columns=["Образование и ВУЗ"], inplace=True)

# Проверка
print(df.head())
```

В результате преобразования у вас должен получиться столбец, содержащий информацию о том, сколько месяцев проработал соискатель.

```python
#ваш код здесь  

import re
import numpy as np

def convert_experience_to_months(text):
    if pd.isnull(text) or text.strip() == "Не указано":
        return np.nan
    
    # Поиск количества лет и месяцев с помощью регулярных выражений
    years = re.search(r"(\d+)\s+год", text)
    months = re.search(r"(\d+)\s+месяц", text)

    total_months = 0

    if years:
        total_months += int(years.group(1)) * 12
    if months:
        total_months += int(months.group(1))
    
    return total_months
```

    Еще один важный факт: при выгрузки данных у некоторых соискателей "потерялась" информация о готовности к командировкам. Давайте по умолчанию будем считать, что такие соискатели не готовы к командировкам.
    
Выполните преобразования и удалите столбец **"Город, переезд, командировки"** из таблицы.

*Совет: обратите внимание на то, что структура текста может меняться в зависимости от указания ближайшего метро. Учите это, если будете использовать порядок слов в своей программе.*


```python
#ваш код здесь
million_cities = ['Новосибирск', 'Екатеринбург','Нижний Новгород','Казань',
                  'Челябинск','Омск', 'Самара', 'Ростов-на-Дону', 'Уфа',
                  'Красноярск', 'Пермь', 'Воронеж','Волгоград']

import numpy as np

def parse_location_info(value):
    if pd.isnull(value):
        return pd.Series([np.nan, False, False])  # если пропущено — пустой город, без переезда и командировок

    parts = [part.strip() for part in value.split(',')]

    # --- Определение города ---
    city = parts[0]
    if city in ['Москва', 'Санкт-Петербург']:
        city_cat = city
    elif city in million_cities:
        city_cat = 'город-миллионник'
    else:
        city_cat = 'другие'

    # --- Готовность к переезду ---
    text = value.lower()
    if 'не готов' in text or 'не готова' in text:
        rel = False
    elif 'хочу переехать' in text or 'готова к переезду' in text or 'готов к переезду' in text:
        rel = True
    else:
        rel = False  # если нет инфы — считаем, что не готов

    # --- Готовность к командировкам ---
    if 'не готов к командировкам' in text or 'не готова к командировкам' in text:
        travel = False
    elif 'готов к командировкам' in text or 'готова к командировкам' in text:
        travel = True
    elif 'командировк' in text:  # например, "готова к редким командировкам"
        travel = True
    else:
        travel = False  # если совсем не указано — считаем, что не готов

    return pd.Series([city_cat, rel, travel])

```

```python
df[["Город", "Готовность к переезду", "Готовность к командировкам"]] = df["Город, переезд, командировки"].apply(parse_location_info)
df.drop(columns=["Город, переезд, командировки"], inplace=True)
```

Давайте создадим признаки-мигалки для каждой категории: если категория присутствует в списке желаемых соискателем, то в столбце на месте строки рассматриваемого соискателя ставится True, иначе - False.

Такой метод преобразования категориальных признаков называется One Hot Encoding и его схема представлена на рисунке ниже:
<img src=https://raw.githubusercontent.com/AndreyRysistov/DatasetsForPandas/main/ohe.jpg>
Выполните данное преобразование для признаков "Занятость" и "График", ответьте на контрольные вопросы, после чего удалите их из таблицы

```python
#ваш код здесь
employment_types = ['полная занятость', 'частичная занятость', 'проектная работа', 'волонтерство', 'стажировка']
schedule_types = ['полный день', 'сменный график', 'гибкий график', 'удаленная работа', 'вахтовый метод']
```

```python
def create_ohe_columns(df, column_name, categories):
    for category in categories:
        df[category] = df[column_name].apply(
            lambda x: category in x if isinstance(x, str) else False
        )
    return df

# Обрабатываем 'Занятость'
employment_types = ['полная занятость', 'частичная занятость', 'проектная работа', 'волонтерство', 'стажировка']
df = create_ohe_columns(df, 'Занятость', employment_types)

# Обрабатываем 'График'
schedule_types = ['полный день', 'сменный график', 'гибкий график', 'удаленная работа', 'вахтовый метод']
df = create_ohe_columns(df, 'График', schedule_types)

```

4. Умножить сумму желаемой заработной платы на присоединенный курс валюты (close) и разделить на пропорцию (обратите внимание на пропуски после объединения в этих столбцах), результат занести в новый столбец "ЗП (руб)".


```python
#ваш код здесь
rates = pd.read_csv("currency_rates.csv")
rates['date'] = pd.to_datetime(rates['date'])

currency_mapping = {
    'руб.': 'RUR',
    'USD': 'USD',
    'EUR': 'EUR',
    'KGS': 'KGS',
    'UZS': 'UZS',
    'KZT': 'KZT',
    'грн.': 'UAH',
    'бел. руб.': 'BYN',
    'сум': 'UZS',
}


df['ЗП_число'] = df['ЗП'].str.extract(r'(\d+)').astype(float)
df['ЗП_валюта'] = df['ЗП'].str.extract(r'([а-яА-Я.]+|USD|EUR|KZT|KGS|UZS)')
df['ЗП_валюта'] = df['ЗП_валюта'].map(currency_mapping)

df['Дата_обновления'] = pd.to_datetime(df['Обновление резюме']).dt.date
```

```python
def convert_salary(row):
    val = row['ЗП_число']
    curr = row['ЗП_валюта']
    date = pd.to_datetime(row['Дата_обновления'])
    
    if pd.isna(val) or pd.isna(curr):
        return np.nan
    
    if curr == 'RUR':
        return val
    
    rate_row = rates[(rates['currency'] == curr) & (rates['date'] == date)]
    
    if rate_row.empty:
        return np.nan
    
    proportion = rate_row['proportion'].values[0]
    close = rate_row['close'].values[0]
    
    return val * (close / proportion)
```

```python
df['ЗП (руб)'] = df.apply(convert_salary, axis=1)
df.drop(columns=['ЗП', 'ЗП_число', 'ЗП_валюта', 'Дата_обновления'], inplace=True)
```

```python
# ваш код здесь
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Логарифмируем возраст (чтобы не было проблем с нулями, добавим 1)
log_age = np.log(df['Возраст'] + 1)

# Рассчитаем среднее и стандартное отклонение логарифма возраста
mean_log_age = log_age.mean()
std_log_age = log_age.std()

# Построим гистограмму в логарифмическом масштабе
plt.figure(figsize=(10,6))
sns.histplot(log_age, bins=30, kde=False, color='skyblue')

# Добавим линии: среднее и границы интервала метода 3 сигм
plt.axvline(mean_log_age, color='k', lw=2, label='Среднее')
plt.axvline(mean_log_age - 3*std_log_age, color='r', lw=2, linestyle='--', label='-3σ')
plt.axvline(mean_log_age + 4*std_log_age, color='r', lw=2, linestyle='--', label='+4σ (порог выброса)')

plt.title('Распределение возраста в логарифмическом масштабе')
plt.xlabel('log(Возраст + 1)')
plt.ylabel('Количество')
plt.legend()

# Комментарий по асимметрии:
# Распределение асимметрично вправо, с длинным правым хвостом.

plt.show()

# Поиск выбросов по z-отклонению с порогом 4 сигмы справа
z_scores = (log_age - mean_log_age) / std_log_age
outliers = df[z_scores > 4]

print("Выбросы по возрасту (логарифмический масштаб, > +4σ):")
print(outliers[['Возраст']])

# Удаляем выбросы из основного датафрейма
df = df[z_scores <= 4]
```
